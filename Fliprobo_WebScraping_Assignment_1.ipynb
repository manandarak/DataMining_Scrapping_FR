```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

# Create a new instance of the Chrome driver
driver = webdriver.Chrome()

# Go to the IMDb Top 250 page
driver.get('https://www.imdb.com/chart/top/')

# Initialize lists to store the titles and years
titles = []
years = []

# Locate the movie containers
containers = driver.find_elements(By.CSS_SELECTOR, '[data-testid="chart-layout-main-column"]')

# Loop through each container and extract the data
for container in containers:
    try:
        # Extract the title
        title = container.find_element(By.CLASS_NAME , "ipc-title__text").text
        
        # Extract the year
        year = container.find_element(By.XPATH, '//div[@class="sc-b189961a-7 feoqjK cli-title-metadata"]/span[@class="sc-b189961a-8 kLaxqf cli-title-metadata-item"]').text
        
        titles.append(title)
        years.append(year)
    
    except NoSuchElementException as e:
        print("Exception raised: ", e)

# Close the WebDriver
driver.quit()

# Create a DataFrame from the scraped data
data = {
    'Titles': titles,
    'Years': years,
}
df = pd.DataFrame(data)

df
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
page = requests.get ('https://www.presidentofindia.gov.in/index.php/former-presidents')
soup = BeautifulSoup(page.content,'html.parser')
header_tages =  soup.find_all('div',class_= 'desc-sec')

presidents = []
for president in header_tages:
    name = president.find('h3').text.strip()
    term = president.find('h5').text.strip()
    presidents.append({'Name': name, 'Term': term})

# Create a Pandas DataFrame from the extracted data
df = pd.DataFrame(presidents)

# Display the DataFrame
df
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

# Set up the WebDriver
driver = webdriver.Chrome()

# Open the ICC team rankings page
url = 'https://www.icc-cricket.com/rankings/team-rankings/mens/odi'
driver.get(url)

# Wait for the table to load
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, 'si-body'))
)

# Lists to store the scraped data
team_names = []
matches = []
points = []
ratings = []

# Locate the table rows containing the team data
rows = driver.find_elements(By.XPATH, '//div[@class="si-table-row"]')

# Loop through each row and extract the data
for row in rows:
    try:
        team_name = row.find_element(By.XPATH, './/h3[@class="si-team-name"]/span[@class="si-fname si-text"]').text
        match = row.find_element(By.XPATH, './/div[@class="si-table-data si-matches"]/span[@class="si-text"]').text
        point = row.find_element(By.XPATH, './/div[@class="si-table-data si-pts"]/span[@class="si-text"]').text
        rating = row.find_element(By.XPATH, './/div[@class="si-table-data si-rating"]/span[@class="si-text"]').text
        
        team_names.append(team_name)
        matches.append(match)
        points.append(point)
        ratings.append(rating)
    except NoSuchElementException as e:
        print("Exception raised: ", e)

# Close the WebDriver
driver.quit()

# Create a DataFrame from the scraped data
data = {
    'Team': team_names,
    'Matches': matches,
    'Points': points,
    'Rating': ratings
}
df = pd.DataFrame(data)

df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the webpage content
page = requests.get('https://www.cnbc.com/world/?region=world')
soup = BeautifulSoup(page.content, 'html.parser')

# Extract writers, titles, and times
writers = []
for writer in soup.find_all('span', class_='RiverByline-authorByline'):
    writers.append(writer.text.strip())

titles = []
for title in soup.find_all('div', class_='RiverHeadline-headline RiverHeadline-hasThumbnail'):
    titles.append(title.text.strip())

times = []
for time in soup.find_all('span', class_='RiverByline-datePublished'):
    times.append(time.text.strip())

# # Create a DataFrame
# df = pd.DataFrame({
#     'Writer': writers,
#     'Title': titles,
#     'Time': times
# })

# # Print the DataFrame to verify
# print(df)

# Ensure all lists are the same length by filling missing values with None
max_len = max(len(writers), len(titles), len(times))
writers.extend([None] * (max_len - len(writers)))
titles.extend([None] * (max_len - len(titles)))
times.extend([None] * (max_len - len(times)))

df = pd.DataFrame({
    'Writer': writers,
    'Title': titles,
    'Time': times
})

df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the webpage content
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup = BeautifulSoup(page.content, 'html.parser')


restaurant_names = []
for restaurant in soup.find_all('a', class_='restnt-name ellipsis'):
    restaurant_names.append(restaurant.text)
locations = []
for location in soup.find_all('div', class_='restnt-loc ellipsis'):
    locations.append(location.text)
ratings = []
for rating in soup.find_all('div', class_='restnt-rating rating-4'):
    ratings.append(rating.text)
images = []
for image in soup.find_all('img', class_='no-img'):
    images.append(image['data-src'])
cuisines = []
for cuisine in soup.find_all('a', attrs={'data-w-onclick': 'stopClickPropagation|w1-restarant'}):
    cuisines.append(cuisine.text)

# Ensure all lists are the same length by filling missing values with None ##GOOGLE
max_len = max(len(restaurant_names), len(locations), len(ratings), len(images), len(cuisines))
restaurant_names.extend([None] * (max_len - len(restaurant_names)))
locations.extend([None] * (max_len - len(locations)))
ratings.extend([None] * (max_len - len(ratings)))
images.extend([None] * (max_len - len(images)))
cuisines.extend([None] * (max_len - len(cuisines)))


df = pd.DataFrame({
    'Restaurant Name': restaurant_names,
    'Location': locations,
    'Rating': ratings,
    'Image': images,
    'Cuisine': cuisines
})

df.head(6)
```
